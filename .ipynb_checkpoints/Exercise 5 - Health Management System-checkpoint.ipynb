```python
import datetime
def gettime():
    return datetime.datetime.now()

def log(k):
    if (k==1):
        c= int(input("Enter 1 for exercise and 2 for food : "))
        if(c==1):
            value=input("type here\n")
            with open("Harry-ex.txt","a") as f:
                f.write(str([str(gettime())]) +" : "+ value + "\n")
            print("Successfully written")
        elif(c==2):
            value=input("type here\n")
            with open("Harry-food.txt","a") as f:
                f.write(str([str(gettime())]) +" : "+ value + "\n")
            print("Successfully written")
        
        elif (k==2):
            c= int(input("Enter 1 for exercise and 2 for food : "))
            if(c==1):
                value=input("type here\n")
                with open("Rohan-ex.txt","a") as f:
                    f.write(str([str(gettime())]) +" : "+ value + "\n")
                print("Successfully written")
            elif(c==2):
                value=input("type here\n")
                with open("Rohan-food.txt","a") as f:
                    f.write(str([str(gettime())]) +" : "+ value + "\n")
                print("Successfully written")
                
        elif (k==3):
            c= int(input("Enter 1 for exercise and 2 for food : "))
            if(c==1):
                value=input("type here\n")
                with open("Hammad-ex.txt","a") as f:
                    f.write(str([str(gettime())]) +" : "+ value + "\n")
                print("Successfully written")
            elif(c==2):
                value=input("type here\n")
                with open("Hammad-food.txt","a") as f:
                    f.write(str([str(gettime())]) +" : "+ value + "\n")
                print("Successfully written")
            
        else:
            print("Please enter valid input(1(Harry),2(Rohan),3(Hammad))")
            
def retrieve(k):
    if k==1:
        c= int(input("Enter 1 for exercise and 2 for food : "))
        if(c==1):
            with open("Harry-ex.txt","a") as f:
                for i in f:
                    print(i,end="")
        elif(c==2):
            with open("Harry-food.txt","a") as f:
                for i in f:
                    print(i,end="")
                    
    elif k==2:
        c= int(input("Enter 1 for exercise and 2 for food : "))
        if(c==1):
            with open("Rohan-ex.txt","a") as f:
                for i in f:
                    print(i,end="")
        elif(c==2):
            with open("Rohan-food.txt","a") as f:
                for i in f:
                    print(i,end="")
                    
    elif k==3:
        c= int(input("Enter 1 for exercise and 2 for food : "))
        if(c==1):
            with open("Hammad-ex.txt","a") as f:
                for i in f:
                    print(i,end="")
        elif(c==2):
            with open("Hammad-food.txt","a") as f:
                for i in f:
                    print(i,end="")
                    
    else:
        print("Please enter valid input(1(Harry),2(Rohan),3(Hammad))")
    
    
print("Health Management System : ")
a= int(input("Enter 1 for log the value and 2 for retrieve : "))

if a==1:
    b=int(input("Press 1 for Harry 2 for Rohan 3 for Hammad : "))
    log(b)
else:
    b=int(input("Press 1 for Harry 2 for Rohan 3 for Hammad : "))
    retrieve(b)
```
